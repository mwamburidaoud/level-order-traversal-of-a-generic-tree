```python

# Python3 program to do level order traversal of a generic tree
  
# Represents a node of an n-ary tree
class Node:
     
    def __init__(self, key):
         
        self.key = key
        self.child = []
   
 # Utility function to create a new tree node
def newNode(key):   
    temp = Node(key)
    return temp
     
# Prints the n-ary tree level wise
def LevelOrderTraversal(root):
 
    if (root == None):
        return;
   
    # Standard level order traversal code using queue
    q = []  # Create a queue
    q.append(root); # Enqueue root
    while (len(q) != 0):
     
        n = len(q);
  
        # If this node has children
        while (n > 0):
         
            # Dequeue an item from queue and print it
            p = q[0]
            q.pop(0);
            if p.key in takenClasses:
              continue
          
            else:
              print(p.key, end=' ')
   
            # Enqueue all children of the dequeued item
            for i in range(len(p.child)):
             
                q.append(p.child[i]);
            n -= 1
   
        print() # Print new line between two levels
      
        
        
# Driver program
if __name__=='__main__':

    root = newNode('Math 140');
    (root.child).append(newNode('Math 141'));
    (root.child).append(newNode('CMPSC 121'));
    (root.child).append(newNode('Math 220'));
    (root.child).append(newNode('Phys 211'));
    (root.child[0].child).append(newNode('CMPSC 462'));
    (root.child[0].child[0].child).append(newNode('CMPSC 463'));
    (root.child[1].child).append(newNode('CMPSC 312'));
    (root.child[1].child[0].child).append(newNode('CMPSC 460'));
    (root.child[1].child[0].child).append(newNode('CMPSC 472'));
    (root.child[1].child).append(newNode('CMPSC 122'));
    (root.child[1].child[1].child).append(newNode('CMPSC 221'));
    (root.child[1].child[1].child[0].child).append(newNode('CMPSC 470'));
    (root.child[1].child[1].child[0].child).append(newNode('CMPSC 360'));
    (root.child[1].child[1].child[0].child[0].child).append(newNode('CMPSC 469'));
    

    print ()
    print("Tree Traversal to use for PreReqs")
    LevelOrderTraversal(root);
    
def checkPreReqs(aClass):
    # check to see if class has preReqs
    # if it does, check that they are met/ are in the takenClasses list
    
    # check the path to aClass and if all those classes are in taken, they are good to take it 
    return True
    
    
    
thisSemester = []
futureSemesters = []
    
for aClass in missingClasses:
    if checkPreReqs(aClass): # if true - preReqs met
        thisSemester.append(aClass)
    else:
        futureSemesters.append(aClass)
        
print ()
print ("This Semester: ", thisSemester)
print ()
print ("Future Semesters: ", futureSemesters)

    
#sheet_url = "https://docs.google.com/spreadsheets/d/18KHmSPRHIfliuB87DqkX10bxAdguhyIfGz1OsmTUYyE/edit#gid=0"
#url_l = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=') 
#pd.read_csv(url_l)




# for each class in missing list
# check if it can be taken 
# create/ call a function to verify if prereqs are met
# function checks the tree to see if classes leading up to the class are in the taken list
# if prereqs are met, add to next semester list
# else, add to the future semester list


# "C:\Users\ggind\Downloads\CMPSC_Template.xlsx - CMPSC (2).csv"


________________________________________

Class graph to find paths: 

class Graph:
  def __init__(self, edges):
    self.edges = edges
    self.graph_dict = {}
    for start, end in self.edges:
      if start in self.graph_dict:
        self.graph_dict[start].append(end)
      else:
        self.graph_dict[start] = [end]

    print("Graph dict: ", self.graph_dict)

  def get_paths(self, start, end, path=[]):
    path = path + [start]
  
    if start == end:
      return [path]

    if start not in self.graph_dict:
      return []

    paths = []

    for node in self.graph_dict[start]:
      if node not in path:
        new_paths = self.get_paths(node, end, path)
        for p in new_paths:
          paths.append(p)

    return paths

  def print_path(self):
    for key, val in Graph.items():
      print(f"{key}-->{val}" + '\n')
  

if __name__ == '__main__':
  routes = [
    ("Math 40", "Math 140"),
    ("Math 140", "Phys 211"),
    ("Math 140", "CMPSC 121"),
    ("Math 140", "Math 141"),
    ("Math 140", "Math 220"),
    ("CMPSC 121", "CMPSC 122")
]

routes_graph = Graph(routes)

start = "Math 40"
end = "Math 220"

print(f"Paths between {start} and {end}: ",routes_graph.get_paths(start, end))


```
